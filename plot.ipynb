```python
import os
import numpy as np
from scipy.io import loadmat, savemat
import matplotlib.pyplot as plt
```

```python
rootdir = 'log'
subdir = '2019-02-08--13:21:55'

data = loadmat(os.path.join(rootdir, subdir, "logs.mat"))
returns = data["returns"][0]

plt.plot(returns)
plt.title("Performance")
plt.xlabel("Iteration number")
plt.ylabel("Return")
plt.show()
```
